```python
# Location type analysis
lg = train_data.groupby(['location_type', 'bank_account']) ['bank_account'].count()

rpy = ((lg['Rural']['Yes'] / (lg['Rural']['Yes'] + lg['Rural']['No'])) * 100).round(2)
rpn = ((lg['Rural']['No'] / (lg['Rural']['Yes'] + lg['Rural']['No'])) * 100).round(2)

upy = ((lg['Urban']['Yes'] / (lg['Urban']['Yes'] + lg['Urban']['No'])) * 100).round(2)
upn = ((lg['Urban']['No'] / (lg['Urban']['Yes'] + lg['Urban']['No'])) * 100).round(2)


location_df = pd.DataFrame(data=[[rpy, rpn], [upy, upn]], columns=['% with account', '% without account'], index=['Rural', 'Urban'])
location_df
```
