```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
train_data_path = os.getcwd() + '/train.csv'
test_data_path = os.getcwd() + '/test.csv'
variable_def_path = os.getcwd() + '/variable_definitions.csv'
```

```python
# Retrieving the train data
train_data = pd.read_csv(train_data_path)

# View the train data
train_data.head()
```

```python
# Variable definitions
variable_def = pd.read_csv(variable_def_path)
variable_def
```

```python
# Country analysis
# Number of interviewees in each country
num_of_interviewees = train_data.groupby(['country']) ['country'].count()
num_of_interviewees.plot.bar()
plt.show()

# Number of interviewees with bank accounts according to country
with_account = train_data.groupby(['country', 'bank_account']) ['bank_account'].count()
with_account.unstack().plot.bar()
plt.show()
```

```python
# Year analysis
# Number of interviewees interviewed each year
num_of_interviewees = train_data.groupby(['year']) ['year'].count()
num_of_interviewees.plot.bar()

# Number of interviewees with bank accounts according to year of interview
with_account = train_data.groupby(['year', 'bank_account']) ['bank_account'].count()
with_account.unstack().plot.bar()
plt.show()
```

```python
# Location type analysis
# Number of interviewees in each loaction type
num_of_interviewees = train_data.groupby(['location_type']) ['location_type'].count()
num_of_interviewees.plot.bar()

# Number of interviewees with bank accounts according to location type
with_account = train_data.groupby(['location_type', 'bank_account']) ['bank_account'].count()
with_account.unstack().plot.bar()
plt.show()
```

```python
# Cellpnone access analysis
# Number of interviewees with/without cellphone access
num_of_interviewees = train_data.groupby(['cellphone_access']) ['cellphone_access'].count()
num_of_interviewees.plot.bar()

# Number of interviewees with bank accounts according to cellphone access
with_account = train_data.groupby(['cellphone_access', 'bank_account']) ['bank_account'].count()
with_account.unstack().plot.bar()
plt.show()
```

```python
# Household size analysis
# Number of interviewees with differnt household sizes
num_of_interviewees = train_data.groupby(['household_size']) ['household_size'].count()
num_of_interviewees.plot.bar()

# Number of interviewees with bank accounts according to cthe household size
with_account = train_data.groupby(['household_size', 'bank_account']) ['bank_account'].count()
with_account.unstack().plot.bar()
plt.show()
```
