```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
train_data_path = os.getcwd() + '/train.csv'
variable_def_path = os.getcwd() + '/variable_definitions.csv'
```

```python
# Retrieving the train data
train_data = pd.read_csv(train_data_path)

# View the train data
train_data.head()
```

```python
# Variable definitions
variable_def = pd.read_csv(variable_def_path)
variable_def
```

```python
# Country analysis
# Number of interviewees with bank accounts according to country
with_account = train_data.groupby(['country', 'bank_account']) ['bank_account'].count()
with_account.unstack().plot.bar()
plt.show()
```

```python
# Year analysis
# Number of interviewees with bank accounts according to year of interview
with_account = train_data.groupby(['year', 'bank_account']) ['bank_account'].count()
with_account.unstack().plot.bar()
plt.show()
```

```python
# Location type analysis
# Number of interviewees with bank accounts according to location type
with_account = train_data.groupby(['location_type', 'bank_account']) ['bank_account'].count()
with_account.unstack().plot.bar()
plt.show()
```

```python
# Cellpnone access analysis
# Number of interviewees with bank accounts according to cellphone access
with_account = train_data.groupby(['cellphone_access', 'bank_account']) ['bank_account'].count()
with_account.unstack().plot.bar()
plt.show()
```

```python
# Categorizing interviewees according to the number of the household
def household_category(size):
    if size <= 5:
        return 'Xsmall'
    elif size > 5 and size <= 9:
        return 'Small'
    elif size > 9 and size <= 15:
        return 'Medium'
    elif size > 15:
        return 'Large'
    
train_data['household_size'] = train_data['household_size'].apply(household_category)
```

```python
# Household size analysis
# Number of interviewees with bank accounts according to the household category
with_account = train_data.groupby(['household_size', 'bank_account']) ['bank_account'].count()
with_account.unstack().plot.bar()
plt.show()
```

```python
# Categorizing interviewees according to age
def age_category(age):
    if age <= 17:
        return 'Teen'
    elif age > 17 and age <= 40:
        return 'Youth'
    elif age > 40 and age <= 60:
        return 'Adult'
    elif age > 60 and age <= 80:
        return 'Old'
    elif age > 80:
        return 'Very Old'

train_data['age_of_respondent'] = train_data['age_of_respondent'].apply(age_category)
```

```python
# Age analysis
# Number of interviewees with bank accounts according to the age category
with_account = train_data.groupby(['age_of_respondent', 'bank_account']) ['bank_account'].count()
with_account.unstack().plot.bar()
plt.show()
```

```python
# Gender of respondent analysis
# Number of interviewees with bank accounts according to the gender
with_account = train_data.groupby(['gender_of_respondent', 'bank_account']) ['bank_account'].count()
with_account.unstack().plot.bar()
plt.show()
```

```python
# Relationship with head analysis
# Number of interviewees with bank accounts according to the relationship with head
with_account = train_data.groupby(['relationship_with_head', 'bank_account']) ['bank_account'].count()
with_account.unstack().plot.bar()
plt.show()
```

```python
# Marital status analysis
# Number of interviewees with bank accounts according to the marital status
with_account = train_data.groupby(['marital_status', 'bank_account']) ['bank_account'].count()
with_account.unstack().plot.bar()
plt.show()
```

```python
# Education level analysis
# Number of interviewees with bank accounts according to the education level
with_account = train_data.groupby(['education_level', 'bank_account']) ['bank_account'].count()
with_account.unstack().plot.bar()
plt.show()
```

```python
# Job type analysis
# Number of interviewees with bank accounts according to the job type
with_account = train_data.groupby(['job_type', 'bank_account']) ['bank_account'].count()
with_account.unstack().plot.bar()
plt.show()
```
